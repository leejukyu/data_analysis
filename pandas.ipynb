#### 데이터 전처리(pre-processing)

```python
# 판다스 특수문자 제거
# €110.00m -> 110.0
df['value'] = df['value'].str.replace('€','')
df['value'] = df['value'].str.replace('m','').astype('float')
df.head()
```

```python
df.info()
```

#### 칼럼 생성과 삭제

```python
df['시장가치(억)'] = df['value']*13
df.head()
```

```python
#칼럼 삭제
df.drop(columns=['value'], inplace=True)
df.head()
```
